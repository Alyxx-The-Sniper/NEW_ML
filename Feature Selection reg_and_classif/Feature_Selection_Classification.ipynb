```python
# all feature
def train_and_evaluate_classif(X, y, classifier,
                                           test_size=0.3, 
                                           random_state=42
                                          ):
    
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
#         ('selector', SelectKBest(score_func=f_classif, k=2)),
        ('classifier', classifier)
    ])

    # Fit the model
    pipeline.fit(X_train, y_train)
    
            
#     # Normalize the features
#     scaler = StandardScaler()
#     X_train = scaler.fit_transform(X_train)
#     X_test = scaler.transform(X_test)

#     # Initialize and fit the Linear Regression model
#     lr = LogisticRegression(max_iter=1000)
#     lr.fit(X_train, y_train)

    # Evaluate the model
    train_score = pipeline.score(X_train, y_train)
    test_score = pipeline.score(X_test, y_test)

    return train_score, test_score
```

```python
# k best list of scores
def K_best_score_list(score_func):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # normalize
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    selector = SelectKBest(score_func, k='all')
    x_train_kbest = selector.fit_transform(X_train, y_train)
    x_test_kbest = selector.transform(X_test)


    feature_scores = pd.DataFrame({'Feature': X.columns,
                               'Score': selector.scores_,
                               'p-Value': selector.pvalues_})

    feature_scores = feature_scores.sort_values(by='Score', ascending=False)
    return feature_scores
```

```python
# k best evaluate features
def kbest_evaluate_features(X, y,
                      classifier,
                      score_func,
                      test_size=0.3,
                      random_state=42):

    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # Normalize features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    accuracy_list = []
    selected_features_list = []

    for k in range(1, len(X.columns) + 1):
        selector = SelectKBest(score_func, k=k)
        X_train_kbest = selector.fit_transform(X_train_scaled, y_train)
        X_test_kbest = selector.transform(X_test_scaled)

        
########### MODEL ###########################################################        

        classifier = classifier
        classifier.fit(X_train_kbest, y_train)
            

###########################################################################

        y_preds_kbest = classifier.predict(X_test_kbest)
        
        # Calculate the accuracy score
        accuracy_score_kbest = accuracy_score(y_test, y_preds_kbest)
        
        # Optionally calculate AUC score if it's a binary classification
        if len(y.unique()) == 2:
            y_probs_kbest = classifier.predict_proba(X_test_kbest)[:, 1]
            auc_score_kbest = roc_auc_score(y_test, y_probs_kbest)
        else:
            auc_score_kbest = None
        
        accuracy_list.append(accuracy_score_kbest)
        
        # Get selected feature names
        selected_feature_mask = selector.get_support()
        selected_features = X.columns[selected_feature_mask].tolist()
        selected_features_list.append(selected_features)

    x = np.arange(1, len(X.columns) + 1)
    result_df = pd.DataFrame({
        'k': x,
        'accuracy_score': accuracy_list,
        'selected_features': selected_features_list
    })

    return result_df
```

```python
# Recursive Features Elimination
def evaluate_rfe_features(X, y,
                         classifier
                         ):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Normalize features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    accuracy_list = []
    auc_list = []
    selected_features_list = []

    for k in range(1, len(X.columns) + 1):
        #### model ####
        classifier = classifier
        rfe = RFE(estimator=classifier, n_features_to_select=k)
        X_train_rfe = rfe.fit_transform(X_train_scaled, y_train)
        X_test_rfe = rfe.transform(X_test_scaled)
        
        classifier.fit(X_train_rfe, y_train)
        y_preds_rfe = classifier.predict(X_test_rfe)
        
        # Calculate accuracy score
        accuracy_score_rfe = accuracy_score(y_test, y_preds_rfe)
        accuracy_list.append(accuracy_score_rfe)
        
        # Optionally calculate AUC score for binary classification
        if len(y.unique()) == 2:
            y_probs_rfe = classifier.predict_proba(X_test_rfe)[:, 1]
            auc_score_rfe = roc_auc_score(y_test, y_probs_rfe)
        else:
            auc_score_rfe = None
        
        auc_list.append(auc_score_rfe)
        
        # Get selected feature names
        selected_feature_mask = rfe.get_support()
        selected_features = X.columns[selected_feature_mask].tolist()
        selected_features_list.append(selected_features)

    x = np.arange(1, len(X.columns) + 1)
    result_df = pd.DataFrame({
        'k': x,
        'accuracy_score': accuracy_list,
        'auc_score': auc_list,
        'selected_features': selected_features_list
    })

    return result_df
```

```python
# Sequencial Forward and Backward Features Selection
def feature_selection_with_sfs_sbs(X, y,
                                   classifier,
                                   test_size=0.3,
                                   random_state=42,
                                   forward=True,
                                   floating=False,
                                   scoring='accuracy',
                                   cv=5
                                  ):
    # List of feature names
    feature_names = X.columns.tolist()

    # Splitting the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Standardize the data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Initialize lists to store results
    selected_features = []
    scores = []

    # Iterate over different numbers of features
    for k in range(1, X.shape[1] + 1):
        # Initialize the Sequential Feature Selector
        sfs = SFS(classifier,
                  k_features=k,
                  forward=forward,
                  floating=floating,
                  scoring=scoring,
                  cv=cv)
        
        # Fit the Sequential Feature Selector to the training data
        sfs.fit(X_train_scaled, y_train)
        
        # Transform the data to only include the selected features
        X_train_selected = sfs.transform(X_train_scaled)
        X_test_selected = sfs.transform(X_test_scaled)
        
        # Train a new model using only the selected features
        model = classifier
        model.fit(X_train_selected, y_train)
        
        # Evaluate the model on the test set
        y_pred = model.predict(X_test_selected)
        
        if scoring == 'accuracy':
            score = accuracy_score(y_test, y_pred)
        elif scoring == 'roc_auc':
            y_probs = model.predict_proba(X_test_selected)[:, 1]
            score = roc_auc_score(y_test, y_probs)
        else:
            raise ValueError("Unsupported scoring metric. Use 'accuracy' or 'roc_auc'.")
        
        # Store results
        selected_features.append([feature_names[i] for i in sfs.k_feature_idx_])
        scores.append(score)
        
    # Create a DataFrame to store the results
    results_df = pd.DataFrame({
        'Number of Features': list(range(1, X.shape[1] + 1)),
        'Selected Features': selected_features,
        f'{scoring} Score': scores
    })
    
    return results_df
```
